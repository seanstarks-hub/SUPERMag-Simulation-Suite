# Long-Range Spin-Triplet Correlations

In structures with **inhomogeneous magnetization**, conventional spin-singlet
Cooper pairs can be converted into spin-triplet pairs. These triplet
correlations are immune to pair-breaking by the exchange field and can
propagate over **long distances** in ferromagnets — a phenomenon with
deep implications for superconducting spintronics.

## Singlet vs. Triplet Pairing

Cooper pairs carry total spin $S = 0$ (singlet) or $S = 1$ (triplet).
The spin structure determines how pairs interact with magnetic order:

| Pair type | Spin state | $m_s$ | Symbol | Behavior in FM |
|-----------|-----------|-------|--------|----------------|
| Singlet | $\frac{1}{\sqrt{2}}(\!\uparrow\downarrow\rangle - \!\downarrow\uparrow\rangle)$ | 0 | $f_0$ | Oscillates and decays over $\xi_F$ |
| Triplet ($m_s=0$) | $\frac{1}{\sqrt{2}}(\!\uparrow\downarrow\rangle + \!\downarrow\uparrow\rangle)$ | 0 | $f_3$ | Oscillates and decays over $\xi_F$ |
| Triplet ($m_s=+1$) | $\!\uparrow\uparrow\rangle$ | $+1$ | $f_\uparrow$ | **Long-range**: decays over $\xi_N \gg \xi_F$ |
| Triplet ($m_s=-1$) | $\!\downarrow\downarrow\rangle$ | $-1$ | $f_\downarrow$ | **Long-range**: decays over $\xi_N \gg \xi_F$ |

The **equal-spin triplet** components ($m_s = \pm 1$) are special: both electrons
have the same spin, so the exchange field does not break the pair. They decay
over the *normal-metal* coherence length $\xi_N$, which can be orders of
magnitude longer than $\xi_F$.

## Triplet Generation Mechanism

Equal-spin triplets are **not** generated at a simple S/F interface with
uniform magnetization. They require a **spin-mixing** and **spin-rotation**
process, which occurs at interfaces where the magnetization direction changes.

### Two-Step Conversion

1. **Singlet → $m_s = 0$ triplet**: At any S/F interface, the exchange field
   mixes the singlet $f_0$ with the $m_s = 0$ triplet $f_3$. This requires
   only a spin-active interface (always present in S/F).

2. **$m_s = 0$ → $m_s = \pm 1$ triplet**: When the magnetization rotates
   at an interface or domain wall, the $m_s = 0$ component is projected onto
   the $m_s = \pm 1$ basis of the new quantization axis.

The generated equal-spin triplet amplitude is:

$$
f_{\uparrow\uparrow}(x) \propto \sin(\alpha)\, \exp\!\left(-\frac{x}{\xi_N}\right)
$$

where:
- $\alpha$ is the misalignment angle between adjacent magnetization directions
- $\xi_N = \sqrt{\hbar D / (2\pi k_B T)}$ is the normal-metal thermal length
- The amplitude vanishes for collinear ($\alpha = 0$) and is maximal for
  perpendicular ($\alpha = \pi/2$) magnetization

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)  # in units of xi_F
xi_N_over_xi_F = 5.0  # typical ratio for weak FM

# Short-range singlet / m_s=0 triplet
f_singlet = np.exp(-x) * np.cos(x)

# Long-range equal-spin triplet (for alpha = pi/2)
f_triplet = 0.3 * np.exp(-x / xi_N_over_xi_F)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.abs(f_singlet), 'b-', linewidth=2,
        label=r'Singlet $|f_0|$ (decays over $\xi_F$)')
ax.plot(x, f_triplet, 'r-', linewidth=2,
        label=r'Equal-spin triplet $|f_{\uparrow\uparrow}|$ (decays over $\xi_N$)')
ax.axvline(1.0, color='blue', ls=':', alpha=0.5, label=r'$\xi_F$')
ax.axvline(xi_N_over_xi_F, color='red', ls=':', alpha=0.5, label=r'$\xi_N$')
ax.set_xlabel(r'$x / \xi_F$', fontsize=13)
ax.set_ylabel('Pair amplitude', fontsize=13)
ax.set_title('Short-range vs. long-range proximity', fontsize=14)
ax.legend(fontsize=10)
ax.set_ylim(-0.05, 1.05)
fig.tight_layout()
plt.show()
```

## Minimal Structures for Triplet Generation

Several geometries can produce the required magnetization inhomogeneity:

### S/F₁/F₂/F₃ (Bergeret trilayer)
Three ferromagnetic layers with non-collinear magnetization. F₁ and F₃ are
"spin mixers" with magnetization misaligned from F₂ (the central long-range
channel).

### S/F with domain walls
Bloch or Néel domain walls naturally provide continuous magnetization rotation.
The triplet amplitude peaks at the domain wall center.

### S/HM/F (spin-orbit interface)
A heavy-metal interlayer (Pt, W) provides spin-orbit coupling that
acts as an effective spin rotator, converting $m_s = 0$ to $m_s = \pm 1$
without requiring magnetic inhomogeneity.

### Angular Dependence

For a bilayer with misalignment angle $\alpha$:

$$
f_{\uparrow\uparrow} \propto \sin\alpha, \qquad
f_{\downarrow\downarrow} \propto \sin\alpha, \qquad
f_3 \propto \cos\alpha
$$

So the long-range component is maximized at $\alpha = 90°$ and vanishes
at $\alpha = 0°$ (parallel) and $\alpha = 180°$ (antiparallel).

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = np.linspace(0, np.pi, 200)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(np.degrees(alpha), np.sin(alpha)**2, 'r-', linewidth=2,
        label=r'$|f_{\uparrow\uparrow}|^2 \propto \sin^2\alpha$')
ax.plot(np.degrees(alpha), np.cos(alpha)**2, 'b--', linewidth=2,
        label=r'$|f_3|^2 \propto \cos^2\alpha$')
ax.set_xlabel(r'Misalignment angle $\alpha$ (degrees)', fontsize=13)
ax.set_ylabel('Relative amplitude', fontsize=13)
ax.set_title('Triplet components vs. magnetization angle', fontsize=14)
ax.legend(fontsize=11)
ax.set_xlim(0, 180)
ax.set_xticks([0, 45, 90, 135, 180])
fig.tight_layout()
plt.show()
```

## Experimental Evidence

Long-range triplet supercurrents have been observed in several systems:

- **Ho/Co/Ho** (Robinson et al., Science 2010): Supercurrent through ~20 nm Co
  (far exceeding $\xi_F \sim 1$ nm), enabled by conical magnetism in Ho
- **Nb/CrO₂/Nb** (Keizer et al., Nature 2006): Half-metallic CrO₂ is 100%
  spin-polarized — only triplet pairs can carry supercurrent
- **NbTiN/GdN/NbN** (Senapati et al., 2011): Critical current enhanced by
  ferromagnetic insulator GdN acting as spin filter

These experiments confirm the long-range triplet proximity effect and open
pathways toward dissipationless spintronics.
